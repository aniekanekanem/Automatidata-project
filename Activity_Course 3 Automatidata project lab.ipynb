1. Identify any outliers: 


*   What methods are best for identifying outliers?
*   How do you make the decision to keep or exclude outliers from any future models?



Best methods for identifying ouliers:  
- By investigating the mean and median of the concerned data.  Every data is subdvided into outliers, lower limit, 1st quartile, 2nd quartile (median), 3rd quartile, upper limit and then outlier.  Any data away from the lower limit or from the upper limit is an oulier.

- Using the box plot where any data point that falls beyond 1.5 times the inter-quartile range is an outlier.

When to keep or exclude outliers:
- Keep outliers if you are not sure they are errors.
-  The more the data volume, the less likely effect they have on your results.  Hence if your dataset is large enough, you can keep outliers.
- If your dataset is a small datset, for the purpose of retaining as much data as possible, you may want to keep the ouliers, this is to ensure you have enough statistical power.  However, if your dataset happens to contain many outliers, then statistical test may need to be adopted to ensure statistical strength such as using non-parametric statistical tests which performs better for such dataset.

When to remove outliers
- If extreme checks (data exploration, data quality checks, mathematical explanation, etc) have been carried out and proposed removal have been justified, then you can remove it.
- If you have legitimate reasons to do so.  However, every outlier removed must be documented accompanied with reasons for the removal.

```python
#BONUS CONTENT

#1. Generate random points on a 2D plane from a normal distribution
#==> ENTER YOUR CODE HERE
num_points = 100
mean = [0, 0]
cov = [[1, 0], [0, 1]]
points = np.random.multivariate_normal(mean, cov, num_points)

# 2. Calculate Euclidean distances between points in first half and second half of array
#==> ENTER YOUR CODE HERE
halfway = num_points // 2
distances = np.linalg.norm(points[:halfway] - points[halfway:], axis=1)
distances

# 3. Group the coordinates by "drop-off location", compute mean distance
#==> ENTER YOUR CODE HERE
def group_and_mean_distance(points):
    unique_points = np.unique(points, axis=0)
    mean_distances = []
    for unique_point in unique_points:
        indices = np.where((points == unique_point).all(axis=1))
        group = points[indices]
        distances = np.linalg.norm(group - unique_point, axis=1)
        mean_distance = np.mean(distances)
        mean_distances.append((unique_point, mean_distance))
    return mean_distances

# Example usage:
mean_distances = group_and_mean_distance(points)

# 4. Plot the mean distance between each endpoint ("drop-off location") and all points it connected to
#==> ENTER YOUR CODE HERE
for point, mean_distance in mean_distances:
    plt.scatter(points[:, 0], points[:, 1], alpha=0.3)
    plt.scatter(point[0], point[1], color='red', marker='o', label=f"Mean Distance: {mean_distance:.2f}")
    plt.legend()
    plt.show()
```
